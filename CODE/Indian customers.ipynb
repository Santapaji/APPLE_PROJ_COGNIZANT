```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('newmaster.csv')
```

```python
df.head()
```

```python
df['Middle_Name'].fillna('No middle name',inplace=True)
```

```python
df.head()
```

```python
df['Music_Label'].fillna('Others',inplace=True)
```

```python
df['Video_Id'].fillna(0,inplace=True)
```

```python
df=df.assign(Video_Id1=df['Video_Id'])
```

```python
df.loc[df['Video_Id1']!=0,'Video_Id1']=1
```

```python
df['Video_Id1']=df['Video_Id1'].replace(0,'Video_Id not assigned')
```

```python
df['Video_Id1']=df['Video_Id1'].replace(1,'Video_Id assigned')
```

```python
df=df.assign(Age_Category=df['Age'])
```

```python
df.loc[df['Age_Category']>=18,'Age_Category']='18 and above years'
```

```python
df.loc[df['Age_Category']!='18 and above years','Age_Category']='Under 18'
```

```python
sns.set(style='dark')
sns.countplot(x=df['Video_Id1'],hue=df['Age_Category'])
plt.xticks(rotation=30,horizontalalignment='center')
plt.xlabel("Category Of Customer")


plt.title("Customers using video services according to age group")
```

```python
aud=df['Audio_Name'].value_counts()[:6]
```

```python
sns.set(style='dark')
aud.plot(kind='bar',figsize=(10,8))
plt.xticks(rotation=30,horizontalalignment='center')
plt.title("top 6 popular audio tracks among Indian Users")
plt.xlabel("Name of song")
plt.ylabel("Number of count")
```

```python
df['Product_Name']=df['Product_Name'].replace('ipad','Ipad')
df['Product_Name']=df['Product_Name'].replace('macbook','Macbook')
df['Product_Name']=df['Product_Name'].replace('iphone','Iphone')
```

```python
df['Product_Name']=df['Product_Name'].replace('Ipad ','Ipad')
```

```python
music_app=df['Music_App'].value_counts()
```

```python
apple_products=df['Product_Name'].value_counts()
sns.set(style='dark')
apple_products.plot(kind='bar',figsize=(10,8))
plt.xticks(rotation=30,horizontalalignment='center')
plt.title("Popularity of the 3 main apple products among Indian Customers")
plt.xlabel("Apple Product Name")
plt.ylabel("Count")
```

```python
sns.set(style='dark')
music_app.plot(kind='bar',figsize=(10,8))
plt.xticks(rotation=30,horizontalalignment='center')
plt.title("Popularity of music apps among indian users")
plt.xlabel("Name of music app")
plt.ylabel("Number of customers")
```

```python
sns.set(style='dark')
plt.xticks(rotation=30,horizontalalignment='center')
sns.countplot(x=df['Music_App'],hue=df['Product_Name'])
plt.title("Popularity of music apps for customer using apple products ")
plt.xlabel("Music app")
plt.ylabel("Count")
```

```python

music_label=df['Music_Label'].value_counts()
sns.set(style='dark')
music_label.plot(kind='bar',figsize=(10,8))
plt.xticks(rotation=30,horizontalalignment='center')
plt.title("Top music labels preferred among Indian Customers")
plt.xlabel("Name of music label")
plt.ylabel("Count")
```
